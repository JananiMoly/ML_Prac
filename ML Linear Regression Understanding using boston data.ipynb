```python
boston_df = pd.DataFrame(boston_data.data, columns = boston_data.feature_names)
```

```python
boston_df['MEDV']= boston_data.target
```

```python
boston_df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = boston_data.data
y= boston_data.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size =0.2)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LinearRegression
Linereg = LinearRegression()
```

```python
Linereg.fit(X_train, y_train)
```

```python
print("The intercept for this data is",Linereg.intercept_ )
```

```python
print("The coefficient of X is", Linereg.coef_)
```

```python
print("The accuracy score of the trained model is", Linereg.score(X_train, y_train))
```

```python
y_pred = Linereg.predict(X_test)
```

```python
print("Predicting the Y_pred of this model: \n", y_pred)
```

```python
print("predicting the accuracy score of test dataset", Linereg.score(X_test, y_test))
```
